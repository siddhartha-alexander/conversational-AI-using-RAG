```python
from langchain_text_splitters import CharacterTextSplitter
from langchain_chroma import Chroma
from langchain_core.prompts import PromptTemplate
from langchain_core.runnables import RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser
```

```python
from langchain_community.document_loaders import PyPDFLoader
file='/content/NIPS-2017-attention-is-all-you-need-Paper.pdf'
loader=PyPDFLoader(file)
doc=loader.load_and_split()
```

```python
doc[0]
```

```python
len(doc)
```

**Split The Data**

```python
text_split=CharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=100,
    length_function=len
)
```

**Segmentation**

```python
texts=text_split.split_documents(doc)
```

```python
texts[0]
```

```python
len(texts)
```

**embeddings**

```python
from langchain.embeddings import HuggingFaceEmbeddings
embedding_model=HuggingFaceEmbeddings(model_name='all-MiniLM-L6-v2')
```

**DataBase Creation**

```python
vectordb=Chroma(
    collection_name='sid',
    embedding_function=embedding_model
)
```

```python
vectordb
```

**inserting Files into database**

```python
storage_id=vectordb.add_documents(texts)
```

```python
storage_id[0]
```

```python
storage_id[1]
```

**Similarity Search**

```python
res=vectordb.similarity_search(
    query='What problem does the Transformer architecture aim to solve?',
    k=2
)
```

```python
res
```

b.llm

```python
retriver=vectordb.as_retriever()
```

```python
from transformers import AutoTokenizer,AutoModelForSeq2SeqLM,pipeline
```

```python
tokenizer=AutoTokenizer.from_pretrained('google/flan-t5-base')
```

```python
if tokenizer.pad_token is not None:
  tokenizer.add_special_tokens({'pad_token':'[PAD]'})
```

```python
model=AutoModelForSeq2SeqLM.from_pretrained('google/flan-t5-base')
```

```python
model.resize_token_embeddings(len(tokenizer))
```

```python
model.config.pad_token_id=tokenizer.pad_token_id
```

```python
generator=pipeline('text2text-generation',model=model,tokenizer=tokenizer)
```

```python
from langchain.llms import HuggingFacePipeline
llm=HuggingFacePipeline(pipeline=generator)
```

**prompt design**

```python
template="""use the context to answer the questions.If you dont know say i dont know.

         context:
         {context}

         question:
         {question}

          answer"""
```

```python
custom_template=PromptTemplate(
    template=template
)
```

```python
rag_chain=(
    {'context':retriver,'question':RunnablePassthrough()}
    |custom_template
    |llm
    |StrOutputParser()
)
```

**Test**

```python
query='What problem does the Transformer architecture aim to solve?'
res=rag_chain.invoke(query)
res
```

```python
import gradio as gr
def chat(message,history):
  bot_message=rag_chain.invoke(message)
  history.append((message,bot_message))
  return history,history
with gr.Blocks() as demo:
  chatbot=gr.Chatbot()
  msg=gr.Textbox()
  clear=gr.Button('clear')
  msg.submit(chat, [msg,chatbot], [msg,chatbot])
  clear.click(lambda: None, None, chatbot, queue=False)
demo.launch()
```
